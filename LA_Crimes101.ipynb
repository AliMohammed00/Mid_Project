```python
def cl_age(x):
    if x > 10 and x < 21:
        return '5-20'
    elif x > 20 and x < 31:
        return '20-30'
    elif x > 30 and x < 41:
        return '30-40'
    elif x > 40 and x < 61:
        return '40-60'
    elif x > 60 and x < 81:
        return '60-80'
    elif x > 80 and x < 101:
        return '80-100'
    else:
        return 'Other'
```

```python
df['victim_age_cat'] = df['victim_age'].apply(cl_age)
```

```python
def stand_time(x):
    if x >= 0 and x < 400:
        return '12 : 4 AM'
    elif x >= 400 and x < 800:
        return '4 : 8 AM'
    elif x >= 800 and x < 1200:
        return '8 : 12 AM'
    elif x >= 1200 and x < 1600:
        return '12 : 4 PM'
    elif x >= 1600 and x < 2000:
        return '4 : 8 PM'
    elif x >= 2000 and x <= 2400:
        return '8 : 12 PM'
```

```python
df['standard_time'] = df['time_occured'].apply(stand_time)
```

```python
def descent(x):
    if x == 'A':
        return 'Asian'
    elif x == 'B':
        return 'Black'
    elif x == 'H':
        return 'Hispanic/Latin/Mexican'
    elif x == 'W':
        return 'White'
    elif x == 'C':
        return 'Asian'
    elif x == 'D':
        return 'Cambodian'
    elif x == 'F':
        return 'Asian'
    elif x == 'G':
        return 'Guamanian'
    elif x == 'J':
        return 'Asian'
    elif x == 'I':
        return 'American Indian/Alaskan Native'
    elif x == 'K':
        return 'Asian'
    elif x == 'V':
        return 'Asian'
    elif x == 'X':
        return 'Unknown'
    else:
        return 'Other'
```
